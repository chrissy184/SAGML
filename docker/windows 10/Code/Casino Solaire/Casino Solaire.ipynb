```python
%%time
d = pd.to_datetime(data.TradingDate)
d = d.apply(lambda x: str(x).split()[0])
data['TradingYear'], data['TradingMonth'], data['TradingDay'] =  d.str.split("-").str

data['TimeDiff'] = data['TimeStopped'] - data['TimeStarted']
data['CoinDiff'] = data['CoinOut'] - data['CoinIn']


# data['HourCreated'], data['MinCreated'], _ = data['TimeCreated'].str.split(':').str
# data['HourModified'], data['MinModified'], _ = data['TimeModified'].str.split(':').str

data['HourCreated'], _, _ = data['TimeCreated'].str.split(':').str
data['HourModified'], _, _ = data['TimeModified'].str.split(':').str

data['TimeDiffCreateMod']= (pd.to_datetime(data['TimeModified']) - pd.to_datetime(data['TimeCreated'])).\
astype('timedelta64[m]')
```
